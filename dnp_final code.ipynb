```python
# -*- coding: utf-8 -*-
"""dnp_most_updated.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1oMgiqMqhTl1kDYckpDcEwkMeNJGOrpN1
"""

# -*- coding: utf-8 -*-
"""dnp.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1tM__jGabwQZ5kEDtbSH54-xlTxE_D8Ks
"""

# -*- coding: utf-8 -*-
"""
Created on Sat Nov 19 20:39:27 2022

@author: Dhanisht
"""
import heapq
import time 
import numpy as np
import matplotlib.pyplot as plt
import random
edges = []                        #edge servers 
iots=[]                           #iots devices
for i in range(0,4):             
    edges.append("edge" + str(i))                 
    
    
for i in range(0,20):
    iots.append("iot" + str(i+4))
    
print("iots: ",iots)
print("edges: ",edges)

### INPUTS###

avg_pkt_size=8000       #packet size in bits
max_thres=85000*avg_pkt_size # bits 85000*8000 =680000000 bits
                             #maximum load that one edge can take is "max_thres"




frequency=2.4*10**9    # standard frequency of AODV protocol
datarate= 54*10**6     # standard datarate at which data can be transmitted,
                       #unit of datarate bits/sec
# Different Variables , list , dictionary are initialized here                       
extraload = {"edge1" : [], "edge2" : [],"edge3" : [],"edge4" : []}  #extraload on edges, 
                                                                    #exceeding, maximum threshold  
edgelist = ["edge1","edge2","edge3","edge4"]     
remaining_capacity_ofedges = {}                  #remaining space
t0 = 0 # starting time   

print("\n")
print("EDGE1")

####edge1#####
# 4 iots joining to edge1
tjoin_edge1 = [20,30,35,40]   #joining times of iots to edge1 
tsend_edge1 = []               #time taken to send/receive  required data from different iots to EDGE1. 

for i in range(0,4):                   
    
    tsend_edge1.append(random.randint(1,300*10**6)/datarate)    # total time to send/receive  the desired data from different iots
    
print("time taken to send/receive data to edge1 ",tsend_edge1)      
    
tfinal_edge1 = []           #final time when tcp connection stops
for i in range(0,4):
    tfinal_edge1.append(tsend_edge1[i] + tjoin_edge1[i])     
    
load_edge1 = []                      #load due to each iot on edge1

print("final time of iots which are connected to edge1: ",tfinal_edge1)

for i in range(0,4):
    load_edge1.append(int(datarate*(tfinal_edge1[i]-tjoin_edge1[i])))  

print("iots to connect to edge1: ",load_edge1)
      
total_load_edge1 = 0
#summation of all loads contributed by all iots connected to edge1
for i in range(0,4):
    if(total_load_edge1 < max_thres and load_edge1[i] <= (max_thres - total_load_edge1)): #checking threshold condn
      total_load_edge1 += load_edge1[i]
      
    else:
        print("load exceeded at edge1, connecting to some other edge")
        extraload["edge1"].append(load_edge1[i]) 
      
edgeload = total_load_edge1     # total load on edge1    
print("total load on edge1: ",edgeload)    
    
print("extraload: ",extraload)   
print("maximum threshold of edge1",max_thres)
load1 = max_thres - total_load_edge1          #remaining capacity(load taking)  of edge1

print("\n")
print("EDGE2")
####### for edge 2 ###########
# 4 iots joining to edge2
tjoin_edge2 = [25,32,37,48,54]   #joining times of nodes to edge1 
tsend_edge2 = []               #time to send  to edge1 from nodes/iots


for i in range(0,5):
    
    tsend_edge2.append(random.randint(1,300*10**6)/datarate)   #taking randomly generated file size according AODV protocols
    
    

print("time taken to send/receive data to edge2: ",tsend_edge2)    
    

tfinal_edge2 = []                           #final time when tcp connection stops
for i in range(0,5):
    tfinal_edge2.append(tsend_edge2[i] + tjoin_edge2[i])
    
load_edge2 = []                      #load due to each node on edge2

print("final time of iots which are connected to edge2: ",tfinal_edge2)

for i in range(0,5):
    load_edge2.append(int(datarate*(tfinal_edge2[i]-tjoin_edge2[i])))

print("iots to connect to edge2: ",load_edge2)
                    
total_load_edge2 = 0            #summation of all loads contributed by all nodes connected to edge2
for i in range(0,5):
    if(total_load_edge2 < max_thres and load_edge2[i] <= (max_thres - total_load_edge2)):
      total_load_edge2 += load_edge2[i]
      
    else:
        print("load exceeded at edge2, connecting to some other edge")
        extraload["edge2"].append(load_edge2[i]) 
        

print("extraload ",extraload)
edgeload = total_load_edge2
   
    
print("total load on edge2",total_load_edge2)    
print("maximum threshold of edge2: ",max_thres)
load2 = max_thres - total_load_edge2             #remaining capacity(load taking)  of edge2    
print("\n")
print("EDGE3")
##### edge 3  ########
tjoin_edge3 = [27,38,39]   #joining times of nodes to edge3 
tsend_edge3 = []               #time to send  to edge3 from nodes/iots


for i in range(0,3):
    
    tsend_edge3.append(random.randint(1,300*10**6)/datarate)
    
    
print("time taken to send/receive data to edge3:",tsend_edge3)    
    

tfinal_edge3 = []                           #final time when tcp connection stops
for i in range(0,3):
    tfinal_edge3.append(tsend_edge3[i] + tjoin_edge3[i])
    
load_edge3 = []                      #load due to each node on edge3

print("final time of iots which are connected to edge3: ",tfinal_edge3)

for i in range(0,3):
    load_edge3.append(int(datarate*(tfinal_edge3[i]-tjoin_edge3[i])))

print("iots to connect to edge3: ",load_edge3)
               #workload on edge3     


    

total_load_edge3 = 0            #summation of all loads contributed by all nodes connected to edge1
for i in range(0,3):
    if(total_load_edge3 < max_thres and load_edge3[i] <= (max_thres - total_load_edge3)):
      total_load_edge3 += load_edge3[i]
      
    else:
        print("load exceeded on edge , try connceting to some other edge")
        extraload["edge3"].append(load_edge3[i]) 
    

print("extraload ",extraload)
edgeload = total_load_edge3
    
    
print("total load of iots which are connected to edge3: ",total_load_edge3)    
print("maximum threshold of edge3: ",max_thres)
load3 = max_thres - total_load_edge3
print("\n")
print("EDGE4")
###### edge 4 ####

print("edge 4")
tjoin_edge4 = [28,30,37,44,46]   #joining times of nodes to edge4 
tsend_edge4 = []                 #time to send  to edge4 from nodes


for i in range(0,5):
    
    tsend_edge4.append(int(random.randint(1,300*10**6)/datarate))
    
    

print("time taken to send/receive data to edge4: ",tsend_edge4)    

tfinal_edge4 = []                           #final time when tcp connection stops
for i in range(0,5):
    tfinal_edge4.append(tsend_edge4[i] + tjoin_edge4[i])
    
load_edge4 = []                      #load due to each node on edge4

print("final time of iots which are connected to edge4: ",tfinal_edge4)

for i in range(0,5):
    load_edge4.append(int(datarate*(tfinal_edge4[i]-tjoin_edge4[i])))

print("iots to connect to edge4: ",load_edge4)
               #workload on edge4     


    

total_load_edge4 = 0            #summation of all loads contributed by all iots connected to edge1
for i in range(0,5):
    if(total_load_edge4 < max_thres and load_edge4[i] <= (max_thres - total_load_edge4)):
      total_load_edge4 += load_edge4[i]
      
    else:
        print("load exceeded on edge , try connceting to some other edge")
        extraload["edge4"].append(load_edge4[i])               
        
    

print("extraload ",extraload)
edgeload = total_load_edge4
print(edgeload)    
    
print("total load of iots which are connected to edge4: ",total_load_edge4)    
print("maximum threshold of edge4: ",max_thres)
load4 = max_thres - total_load_edge4
print("extraload ",extraload)


for i in edgelist:
  if(i == "edge1"):
    remaining_capacity_ofedges[i] = load1
  if(i == "edge2"):
    remaining_capacity_ofedges[i] = load2
  if(i == "edge3"):
    remaining_capacity_ofedges[i] = load3
  if(i == "edge4"):
    remaining_capacity_ofedges[i] = load4


print("\n")
print("remaining_capcacity_of_loads_on_edges",remaining_capacity_ofedges)


          


# now connecting iots with other edges which have less loads

#edge1 
def edge1():
  print("\n")
  for i in extraload['edge1']:
      
      
      for j in remaining_capacity_ofedges.values():
        if(i <= j):
          
          extraload.update({'edge1':[0]})
          for key,val in remaining_capacity_ofedges.items():
            if (val == j):
              j = j-i
              remaining_capacity_ofedges[key] = j
          
               
  


#edge2
def edge2():
  print("\n")
  
  for i in extraload['edge2']:
      
      
      for j in remaining_capacity_ofedges.values():
        if(i <= j):
          extraload.update({'edge2':[0]})
          
          for key,val in remaining_capacity_ofedges.items():
            if (val == j):
              j = j-i
              remaining_capacity_ofedges[key] = j


#edge3
def edge3():
  print("\n")
  for i in extraload['edge3']:
      
      
      for j in remaining_capacity_ofedges.values():
        if(i <= j):
          extraload.update({'edge3':[0]})
          for key,val in remaining_capacity_ofedges.items():
            if (val == j):
              j = j-i
              remaining_capacity_ofedges[key] = j

            
#edge4
def edge4():
  print("\n")
  for i in extraload['edge4']:
      
      for j in remaining_capacity_ofedges.values():
        if(i <= j):
          extraload.update({'edge4':[0]})
          for key,val in remaining_capacity_ofedges.items():
            if (val == j):
              j = j-i
              remaining_capacity_ofedges[key] = j            

edge1()
edge2()
edge3()
edge4()



print("after_allocating_loads_to_edges_whose_total_load_less_than_threshold: ",extraload) 
print("remaining_capacity_of_edges: ",remaining_capacity_ofedges)


# putting all  iots loads which did not get any edge/server due to exceeding of max_threshold on  edges
extraload_list = []

for i in extraload['edge1']:
  if(i != 0):
   extraload_list.append(i)

for i in extraload['edge2']:
   if(i != 0):
    extraload_list.append(i) 

for i in extraload['edge3']:
   if(i != 0):
    extraload_list.append(i)

for i in extraload['edge4']:
   if(i != 0):
    extraload_list.append(i)  


if(len(extraload_list) == 0):
  print("All iot loads are already balanced")

heapq.heapify(extraload_list)       #implementing min heap for all iots which are not connected to any edges

print("loads of iots which are not connected to any edge(heap_data): ",list(extraload_list))

#edge1 -- when there is extraload , or there is more iots for sending/receving data
if(len(extraload_list)):
  time_edge1 = min(tsend_edge1,default = 0)          #time after which edge1 load will be less than max_thres  
  minpos = 0
  if(len(tsend_edge1)):
   minpos = tsend_edge1.index(min(tsend_edge1))  #to find the index 
   print("time duration after which edge1 load is less than threshold: ",time_edge1)
   print(minpos)
   print(time.perf_counter())  #it starts a random timer 
   time.sleep(time_edge1)
   joining_time_new_iot = tfinal_edge1[minpos]
   print(time.perf_counter())
   new_remaining_load_taking_cap_edge1 = load_edge1[minpos] + remaining_capacity_ofedges['edge1'] 
   remaining_capacity_ofedges.update({'edge1':new_remaining_load_taking_cap_edge1})
   print("capacity_of_edges",remaining_capacity_ofedges)
   del load_edge1[minpos]             # removing iot whose sending/receiving of data is done.  
   del tjoin_edge1[minpos]
   del tfinal_edge1[minpos]
   del tsend_edge1[minpos]
   if(extraload_list[0] <= remaining_capacity_ofedges['edge1']):   #now checking if edge1 have capacity to take iot loads which are in heap 
    new_remaining_cap = remaining_capacity_ofedges['edge1'] - extraload_list[0] 
    remaining_capacity_ofedges.update({'edge1':new_remaining_cap})
    print("remaining_cap_of_edge1_after_extra_load_balance",remaining_capacity_ofedges)
    load_edge1.append(extraload_list[0])   # connecting the waiting iot to edge1
    tjoin_edge1.append(joining_time_new_iot)
    tfinal_edge1.append((extraload_list[0]/datarate) + joining_time_new_iot)
    tsend_edge1.append(extraload_list[0]/datarate)
    heapq.heappop(extraload_list)

if(len(extraload_list)):
 print("extraload_list",list(extraload_list))

#edge2 -- when there is some extraload , or there is more iots for sending/receving data
if(len(extraload_list)):
  time_edge2 = min(tsend_edge2,default = 0)          #time after which edge2 load will be less than max_thres  
  print("time_send_edge2",tsend_edge2)
  minpos = 0
  if(len(tsend_edge2)):
   minpos = tsend_edge2.index(min(tsend_edge2))  #to find the index 
   print(time_edge2)
   print(minpos)
   print(time.perf_counter())  #it starts a random timer 
   time.sleep(time_edge2)
   joining_time_new_iot = tfinal_edge2[minpos]
   print(time.perf_counter())
   new_remaining_load_taking_cap_edge2 = load_edge2[minpos] + remaining_capacity_ofedges['edge2'] 
   remaining_capacity_ofedges.update({'edge2':new_remaining_load_taking_cap_edge2})
   print("capacity_of_edges: ",remaining_capacity_ofedges)
   del load_edge2[minpos]             # removing iot whose sending/receiving of data is done.  
   del tjoin_edge2[minpos]
   del tfinal_edge2[minpos]
   del tsend_edge2[minpos]
   if(extraload_list[0] <= remaining_capacity_ofedges['edge2']):
    new_remaining_cap = remaining_capacity_ofedges['edge2'] - extraload_list[0] 
    remaining_capacity_ofedges.update({'edge2':new_remaining_cap})
    print("capacity_of_edges: ",remaining_capacity_ofedges)
    load_edge2.append(extraload_list[0])   # connecting the waiting iot to edge2
    tjoin_edge2.append(joining_time_new_iot)
    tfinal_edge2.append((extraload_list[0]/datarate) + joining_time_new_iot)
    tsend_edge1.append(extraload_list[0]/datarate)
    heapq.heappop(extraload_list)
    
if(len(extraload_list)):
 print("extraload_list",list(extraload_list))

#edge3 -- when there is extraload , or there is more iots for sending/receving data
if(len(extraload_list)):
  time_edge3 = min(tsend_edge3,default = 0)          #time after which edge1 load will be less than max_thres  
  minpos = 0
  if(len(tsend_edge3)):
   minpos = tsend_edge3.index(min(tsend_edge3))  #to find the index 
   print("time duration after which edge3 load is less than threshold: ",time_edge3)
   print(minpos)
   print(time.perf_counter())  #it starts a random timer 
   time.sleep(time_edge3)
   joining_time_new_iot = tfinal_edge3[minpos]
   print(time.perf_counter())
   new_remaining_load_taking_cap_edge3 = load_edge3[minpos] + remaining_capacity_ofedges['edge3'] 
   remaining_capacity_ofedges.update({'edge3':new_remaining_load_taking_cap_edge3})
   print("capacity_of_edges",remaining_capacity_ofedges)
   del load_edge3[minpos]             # removing iot whose sending/receiving of data is done.  
   del tjoin_edge3[minpos]
   del tfinal_edge3[minpos]
   del tsend_edge3[minpos]
   if(extraload_list[0] <= remaining_capacity_ofedges['edge3']):   #now checking if edge3 have capacity to take iot loads which are in heap 
    new_remaining_cap = remaining_capacity_ofedges['edge3'] - extraload_list[0] 
    remaining_capacity_ofedges.update({'edge3':new_remaining_cap})
    print("remaining_cap_of_edge3_after_extra_load_balance",remaining_capacity_ofedges)
    load_edge3.append(extraload_list[0])   # connecting the waiting iot to edge3
    tjoin_edge3.append(joining_time_new_iot)
    tfinal_edge3.append((extraload_list[0]/datarate) + joining_time_new_iot)
    tsend_edge3.append(extraload_list[0]/datarate)
    heapq.heappop(extraload_list)

if(len(extraload_list)):
 print("extraload_list",list(extraload_list))
#edge4 -- when there is extraload , or there is more iots for sending/receving data
if(len(extraload_list)):
  time_edge4 = min(tsend_edge4,default = 0)          #time after which edge4 load will be less than max_thres  
  minpos = 0
  if(len(tsend_edge4)):
   minpos = tsend_edge4.index(min(tsend_edge4))  #to find the index 
   print("time duration after which edge4 load is less than threshold: ",time_edge4)
   print(minpos)
   print(time.perf_counter())  #it starts a random timer 
   time.sleep(time_edge4)
   joining_time_new_iot = tfinal_edge4[minpos]
   print(time.perf_counter())
   new_remaining_load_taking_cap_edge4 = load_edge4[minpos] + remaining_capacity_ofedges['edge4'] 
   remaining_capacity_ofedges.update({'edge4':new_remaining_load_taking_cap_edge4})
   print("capacity_of_edges",remaining_capacity_ofedges)
   del load_edge4[minpos]             # removing iot whose sending/receiving of data is done.  
   del tjoin_edge4[minpos]
   del tfinal_edge4[minpos]
   del tsend_edge4[minpos]
   if(extraload_list[0] <= remaining_capacity_ofedges['edge4']):   #now checking if edge4 have capacity to take iot loads which are in heap 
    new_remaining_cap = remaining_capacity_ofedges['edge4'] - extraload_list[0] 
    remaining_capacity_ofedges.update({'edge4':new_remaining_cap})
    print("remaining_cap_of_edge1_after_extra_load_balance",remaining_capacity_ofedges)
    load_edge4.append(extraload_list[0])   # connecting the waiting iot to edge3
    tjoin_edge4.append(joining_time_new_iot)
    tfinal_edge4.append((extraload_list[0]/datarate) + joining_time_new_iot)
    tsend_edge4.append(extraload_list[0]/datarate)
    heapq.heappop(extraload_list)

if(len(extraload_list)):
 print("extraload_list",list(extraload_list))

print("iots which wants to connect(inside heap): ",list(extraload_list))
print("remaining_cap_of_edges",remaining_capacity_ofedges)




















```
